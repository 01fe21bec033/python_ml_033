A.Randomly generate x and y using random module and do linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random

```

```python
random.seed(42)
num_samples = 22
x = np.array([random.uniform(1, 50) for _ in range(num_samples)]).reshape((-1, 1))
y = np.array([2 * val + random.uniform(-5, 5) for val in x])
```

```python
plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)


mse = mean_squared_error(y, pred)
plt.plot(x, pred, color='green', marker='*')


print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)

```

B.Read an excel file to fetch x and y values and repeat linear regression.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

```

```python
data = pd.read_excel('/content/drive/MyDrive/task2 (1).xlsx')
x = data['x'].values.reshape(-1, 1)
y = data['y'].values
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
```

```python
plt.scatter(x, y, label='Data')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()

print("Slope:", slope)
print("Intercept:", intercept)
```
